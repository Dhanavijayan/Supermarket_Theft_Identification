```python
import pandas as pd
import numpy as np

file_path = "supermarket_sample.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import math

df['Theft_Level'] = df['Theft_Level'].apply(
    lambda x: 1 if x == 'High' else 0
)

X = df[['Shrinkage_INR']]
y = df['Theft_Level']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.8, random_state=42
)

model = LogisticRegression()
model.fit(X_train, y_train)

y_predicted = model.predict(X_test)

print("Accuracy:", model.score(X_test, y_test))

print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def prediction_function(shrinkage):
    z = model.coef_[0][0] * shrinkage + model.intercept_[0]
    return sigmoid(z)

print(prediction_function(200))
print(prediction_function(500))
```

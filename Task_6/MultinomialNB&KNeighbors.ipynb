```python
import pandas as pd
df = pd.read_csv("supermarket_sample.csv")
df.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

model = MultinomialNB()
v = CountVectorizer()

features = ['Foot_Traffic', 'Staff_Count', 'Hour', 'High_Value_Items',
            'Shrinkage_INR', 'Predicted_INR', 'Residual_INR']
X = df[features]
y = df['Theft_Level']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)

X_train_text = X_train.astype(str).agg(' '.join, axis=1)
X_test_text  = X_test.astype(str).agg(' '.join, axis=1)

X_train_vec = v.fit_transform(X_train_text)
X_test_vec  = v.transform(X_test_text)

model.fit(X_train_vec, y_train)
print('Test score:', model.score(X_test_vec, y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=10)
model_knn.fit(X_train_vec.toarray(), y_train)
print('KNN score:', model_knn.score(X_test_vec.toarray(), y_test))
```
